## Defining Model

Lets say it has single neuron in single layer(perceptron):

we have 30 columns in X_train so it must have 30 weights and 1 bias


```python
class MySimpleNN():
  def __init__(self,X):
    self.weights =torch.rand(X.shape[1],1,dtype=torch.float64,requires_grad=True)
    self.bias = torch.zeros(1,dtype=torch.float64,requires_grad=True)

  def forward(self,X):
    z= torch.matmul(X,self.weights)+ self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  def loss_funtion(self,y_pred,y):
    #clamp predictions to avoid log(0)
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred,epsilon,1-epsilon)
    #Note y_pred is 2d tensor but y is y_train_tensor ie 1d tensor so make it 2 dim by adding 1 dim
    y=y.view(-1,1) #in view we can use -1 to infer the exact shape and 1 to add dim
    #calculate loss
    loss = -(y *torch.log(y_pred) + (1-y)* torch.log(1-y_pred)).mean()
    
    return loss

```

## Important Parameters


```python
learning_rate = 0.1
epochs = 25
```

## Training Pipeline

```python

# create model
model = MySimpleNN(X_train_tensor)
# define loop
for epoch in range(epochs):
  # forward pass
  y_pred = model.forward(X_train_tensor)
  print("y_pred",y_pred)
  

  # loss calculate
  loss = model.loss_funtion(y_pred,y_train_tensor)
  print("loss",loss)
  
  # Backward pass
  loss.backward()

  # parameters update
  with torch.no_grad():  # we dont need to track this for gradient
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad

  #zero gradients
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  #print loss in each epoch
  print(f'Epoch: {epoch+1},Loss: {loss.item()}')

```

## Evaluation

```python
with torch.no_grad():
  y_pred = model.forward(X_test_tensor)

#say threshold is 0.5
y_pred = (y_pred > 0.5).float()
accuracy = (y_pred == y_test_tensor).float().mean()
print(f'Accuracy: {accuracy.item()}')

```
