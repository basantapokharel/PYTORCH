## Train Test Split

```python
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2,random_state=32)
```

## Defining Dataset and Dataloader class

```python
from torch.utils.data import Dataset,DataLoader

class CustomDataset(Dataset):
    def __init__(self,features,labels):
        self.features = features
        self.labels = labels
    
    def __len__(self):
        return len(self.features)

    def __getitem__(self,idx):
        return self.features[idx],self.labels[idx]
    
```

```python
train_dataset = CustomDataset(X_train_tensor,y_train_tensor)
test_dataset = CustomDataset(X_test_tensor,y_test_tensor)
```

```python
#dataloader
train_dataloader = DataLoader(train_dataset,batch_size=32,shuffle=True)
test_dataloader = DataLoader(test_dataset,batch_size=32,shuffle=True)
```

## Defining Model

Lets say it has single neuron in single layer(perceptron):

we have 30 columns in X_train so it must have 30 weights and 1 bias


```python
import torch.nn as nn
```

```python
class MySimpleNN(nn.Module):
    def __init__(self,num_features):
        super().__init__()
        self.network =nn.Sequential(
            nn.Linear(num_features,1),
            nn.Sigmoid()
        )
    def forward(self,features):
        y_pred = self.network(features)
        return y_pred

    
```

## Important Parameters


```python
learning_rate = 0.1
epochs = 25
```

## Training Pipeline

```python

# create model
model = MySimpleNN(X_train_tensor.shape[1])

#define loss function
loss_function = nn.BCELoss()  #Binary Cross Entropy Loss

#define optimizer here sgd
optimizer = torch.optim.SGD(model.parameters(),lr=learning_rate)

# define loop
for epoch in range(epochs):
  for batch_features,batch_labels in train_dataloader:
    #since nn.Linear only works for float32 here X_train_tensor is in float64 so gives error so change it
    batch_features = batch_features.float()
    #forward pass
    y_pred = model(batch_features)
  
    # loss calculate
    # y_pred is 2d tensor and batch_labels is 1d tensor so we need to reshape it otherwise it can give diff dim error
    loss = loss_function(y_pred,batch_labels.view(-1,1))

    #zero gradients
    optimizer.zero_grad()

    # Backward pass
    loss.backward()

    #parameters update
    optimizer.step()

    #print loss in each epoch
  print(f'Epoch: {epoch+1},Loss: {loss.item()}')

```

## Evaluation

model.eval() tells PyTorch that the model is in inference mode, which:

Disables Dropout → Ensures neurons are not randomly dropped.
Freezes Batch Normalization statistics → Uses running mean/variance instead of mini-batch stats.

```python
# model evaluation using test_loader
model.eval()  #set the model to evaluation mode removes dropouts and batch_norm
accuracy_list=[]
with torch.no_grad():
  for batch_features,batch_labels in test_dataloader:
    batch_features = batch_features.float()
    #forward pass
    y_pred = model.forward(batch_features)
    y_pred = (y_pred >0.5).float() #convert to 0 and 1

    #compute accuracy
    batch_accuracy = (y_pred == batch_labels.view(-1,1)).float().mean().item()
    accuracy_list.append(batch_accuracy)



#print overall accuracy
overall_accuracy = sum(accuracy_list)/len(accuracy_list)
print(f'Accuracy: {overall_accuracy:.4f}')

```
